```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
nb_maps = 3
nb_algo = 3
run_times = np.transpose(np.array([[0.0708, 0, 0], [0.03595, 0, 0], [0.0080, 0.7569, 1.8549]]))
scores = np.transpose(np.array([[521, 0, 0], [521, 0, 0], [526, 539, 540]]))
expendedNodes = np.transpose(np.array([[598, 0, 0], [342, 0, 0], [66, 2818, 6033]]))
```

```python
plt.figure(figsize=(16, 10))
plt.title("Score of the different adverserial search algorithms")
plt.xlabel("Maps")
plt.ylabel("Scores")

bar_width = .5
bar_positions = np.arange(nb_maps)*bar_width*(nb_algo+1)

for i in range(nb_algo):
    plt.bar(bar_positions+bar_width*i, scores[:, i], width=bar_width)

plt.xticks(bar_positions+(nb_algo-1)/2*bar_width, ["small map","medium map","large map"])
plt.legend(["minimax" , "alpha-beta pruning", "h-minimax"])
plt.savefig("scores_dumby.png")
```
