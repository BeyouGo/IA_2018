```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
nb_maps = 3
nb_algo = 3
run_times = np.transpose(np.array([[0.0648, 0, 0], [0.0279, 0, 0], [0.0060, 0.6254, 1.3723]]))
scores = np.transpose(np.array([[523, 0, 0], [523, 0, 0], [526, 539, 560]]))
expendedNodes = np.transpose(np.array([[464, 0, 0], [246, 0, 0], [66, 2139, 4153]]))
```

```python
plt.figure(figsize=(16, 10))
plt.title("Running times of different adverserial search algorithms")
plt.xlabel("Maps")
plt.ylabel("Running time (s)")

bar_width = .5
bar_positions = np.arange(nb_maps)*bar_width*(nb_algo+1)

for i in range(nb_algo):
    plt.bar(bar_positions+bar_width*i, run_times[:, i], width=bar_width)

plt.xticks(bar_positions+(nb_algo-1)/2*bar_width, ["Map %d" % i for i in range(nb_maps)])
plt.legend(["Algo %d" % i for i in range(nb_algo)])
plt.savefig("run_time_smarty.png")
```
